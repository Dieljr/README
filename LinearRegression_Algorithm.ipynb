```python
# Carregar pacotes

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sb

import sys

from sklearn.linear_model import LinearRegression
```

```python
# Veja a versão do seu Python
print(sys.version)
```

### Criar massa de dados

```python
X = 1/2 * np.random.rand(400)

y = 5 + 3 * X + np.random.rand(400)
```

```python
# Plotar massa de dados

plt.figure(figsize = (15,12))
sb.scatterplot(X, y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Dsitribuição da massa de dados')
plt.show()
```

```python
# Criar objeto de Linear Regression e ver parametros
# Não iremos alterar parametros do lr, já que não causaria nenhuma efeito siginificativo pelo tipo e tamanho do dataset.
lr = LinearRegression()
lr
```

```python
# Visualização da reta de regressão treinada
plt.figure(figsize = (12,8))
sb.jointplot(X, y, kind = 'reg', color= 'g', truncate = False, height = 7)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# precisa mudar a dimensão, porque é exigencia do algoritmo
X_r = X.reshape(-1,1) 

# Aplicar a regressão linear
lr.fit(X_r,y)
```

### Capturando Atributos

```python
# Veja a aplicação com novos valores de X
# Quando X_teste igual a 0, y = 5,479
# Assim realiza as previsões de novos valores.
X_teste = np.arange(0,10)

plt.plot(lr.intercept_ + X_teste)
plt.xlabel('Valores Novos')
plt.ylabel('Intercepto - Y')
plt.ylim(-1,12)
plt.xlim(-1,12)
plt.show()
```

```python
# Variável independente (intercept_): valor que cruza o eixo y.
# A partir deste valor é que você encontrar os valores de x em um plano cartesiano
# É o valor multiplicador de previsão propriamente dito.
# Veja no gráfico acima

print('y é igual a :',lr.intercept_)
```

```python
# Coeficiente estimado (coef_)
# significa que a cada unidade aumentada de x, y aumenta 2,899.

print(lr.coef_)
```
